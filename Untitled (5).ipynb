* DATA CLEANING

Step 1: Load the Dataset

Read the dataset (bank_data.csv) using pandas.

The dataset URL: https://gitlab.crio.do/me_notebook/me_jupyter_bankattritionanalysis/-/raw/master/bank_attrition_dataset.csv

Skip the metadata row from the dataset.

Display the first few rows to get an overview.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_url = 'https://gitlab.crio.do/me_notebook/me_jupyter_bankattritionanalysis/-/raw/master/bank_attrition_dataset.csv'

df = pd.read_csv(data_url, skiprows=[1], header=[0])
print(df.head())
```

Step 2: Check Column Headers

Print the column names of the dataset.

Standardize the column names: convert to lowercase and replace spaces with underscores.

```python
print(df.columns)
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.columns)
```

Step 3: Handle Missing Values

Check for missing values in the dataset.

Impute or drop based on the type of column:

Salary: Replace missing values with the median.

Balance: Replace missing values with 0 (assume no balance).

Satisfaction Score: Fill missing values with median.

Gender: Drop rows with missing values.

Card Type: Fill with the most frequent value (mode).

```python
print(df.isnull().sum())

df['salary'] = df['salary'].fillna(df['salary'].median())
df['satisfaction_score'] = df['satisfaction_score'].fillna(df['satisfaction_score'].median())
df['balance'] = df['balance'].fillna(0)
df = df.dropna(subset='gender')
df['card_type'] = df['card_type'].fillna(df['card_type'].mode()[0])
print(df.isnull().sum())
print("Number of missing values in a dataset:",df.isnull().sum().sum())
```

Step 4: Clean Up Text Columns

Standardize values in card_type:

Convert all text to uppercase.

Remove any leading/trailing spaces.

```python
df['card_type'].unique()
```

```python
df['card_type'] = df['card_type'].str.upper()
df['card_type'].unique()
```

```python
df['card_type'] = df['card_type'].str.upper().str.strip()
df['card_type'].unique()
```

Step 5: Explore Numeric Columns

Identify all numeric columns.

Count the number of unique values in each.

```python
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()
print(numerical_columns)

number_of_unique_values = {col: df[col].nunique() for col in numerical_columns}
print(number_of_unique_values)
```

Step 6: Outlier Detection and Treatment

Plot boxplots for salary and balance.

Identify and count outliers in salary using IQR method.

Cap salary outliers to the upper bound using np.where.

Plot boxplot for salary distribution after handling outliers

```python
df['salary'].plot(kind='box', vert=False)
plt.title('Salary Distribution')
```

```python
df['balance'].plot(kind='box', vert=False)
plt.title('Balance Distribution')
```

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)

IQR = Q3 - Q1

lower_bound= Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

salary_outliers = df[(df['salary']< lower_bound) | (df['salary']> upper_bound)]
print(salary_outliers[['salary']])

print("Number of outliers in salary:", ((df['salary']< lower_bound) | (df['salary']> upper_bound)).sum())
```

```python
df['salary'] = np.where(df['salary']> upper_bound, upper_bound, df['salary'])

df['salary'].plot(kind='box', vert=False)
plt.title('Salary Distribution - Outliers Handled')
```

Label the axes and display the legend.

```python
df['churn_status'] = df['exited'].map({0: 'Stayed', 1: 'Left'})

grouped = df.groupby(['numofproducts', 'churn_status'])

average_balance_each_group = grouped['balance'].mean().reset_index(name='average').round(2)
print(average_balance_each_group)

ax = sns.barplot(x='numofproducts', y='average', hue='churn_status', data=average_balance_each_group)

for p in ax.patches:
    height = p.get_height()
    ax.text(x=p.get_x() + p.get_width()/2,
           y =height +1,
           s = int(height),
           ha = 'center',
           va='bottom')
    
plt.title('Average Balance by Number of Products and Churn Status')
plt.xlabel('Number of Products')
plt.ylabel('Average Balance')
plt.show()
```

Filter the dataset to include only rows where hasfd == 1.

Use .value_counts(normalize=True) on churn_status, rounded to two decimal places.

Create a pie chart using these proportions.

Title the chart "Churn Distribution Among FD Holders".
